```python
from tensorflow.keras.datasets import imdb
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

from tensorflow.keras import regularizers
```

```python
(train_data, train_labels), _ = imdb.load_data(num_words=10000)

```

```python
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

train_data = vectorize_sequences(train_data)
```

```python
model = keras.Sequential([layers.Dense(16, activation="relu"),
                          layers.Dense(16, activation="relu"),
                          layers.Dense(1, activation="sigmoid")
                          ])

model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

```

```python
history_original = model.fit(train_data, train_labels,epochs=20, batch_size=512, validation_split=0.4)
```

```python

loss = history_original.history["loss"]
val_loss = history_original.history["val_loss"]
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, "bo", label="Training loss")
plt.plot(epochs, val_loss, "b", label="Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.clf()
acc = history_original.history["accuracy"]
val_acc = history_original.history["val_accuracy"]
plt.plot(epochs, acc, "bo", label="Training accuracy")
plt.plot(epochs, val_acc, "b", label="Validation accuracy")
plt.title("Training and validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python

model = keras.Sequential([layers.Dense(16,kernel_regularizer=regularizers.l2(0.002),activation="relu"),
                          layers.Dense(16,kernel_regularizer=regularizers.l2(0.002),activation="relu"),
                          layers.Dense(1, activation="sigmoid")])

model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

```python
history_l2_reg = model.fit(
train_data, train_labels,
epochs=20, batch_size=512, validation_split=0.4)
```

```python
loss = history_l2_reg.history["loss"]
val_loss = history_l2_reg.history["val_loss"]
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, "bo", label="Training loss")
plt.plot(epochs, val_loss, "b", label="Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.clf()
acc = history_l2_reg.history["accuracy"]
val_acc = history_l2_reg.history["val_accuracy"]
plt.plot(epochs, acc, "bo", label="Training accuracy")
plt.plot(epochs, val_acc, "b", label="Validation accuracy")
plt.title("Training and validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
# In the preceding listing, l2(0.002) means every coefficient in the weight matrix of
# the layer will add 0.002 * weight_coefficient_value ** 2 to the total loss of the
# model. Note that because this penalty is only added at training time, the loss for this
# model will be much higher at training than at test time
```

#### regulrization: dropout

```python
model = keras.Sequential([layers.Dense(16, activation="relu"),layers.Dropout(0.5),
                          layers.Dense(16, activation="relu"),layers.Dropout(0.5),
                          layers.Dense(1, activation="sigmoid")])

model.compile(optimizer="rmsprop",
loss="binary_crossentropy",
metrics=["accuracy"])

history_dropout = model.fit(train_data, train_labels,
epochs=20, batch_size=512, validation_split=0.4)
```

```python
loss = history_dropout.history["loss"]
val_loss = history_dropout.history["val_loss"]
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, "bo", label="Training loss")
plt.plot(epochs, val_loss, "b", label="Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.clf()
acc = history_dropout.history["accuracy"]
val_acc = history_dropout.history["val_accuracy"]
plt.plot(epochs, acc, "bo", label="Training accuracy")
plt.plot(epochs, val_acc, "b", label="Validation accuracy")
plt.title("Training and validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
